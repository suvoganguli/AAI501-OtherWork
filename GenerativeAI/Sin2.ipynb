```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models

# Function to generate synthetic sinusoidal data
def generate_data(num_samples):
    x = np.linspace(0, 2*np.pi, num_samples)
    y = np.sin(x) + np.random.normal(0, 0.1, size=x.shape)
    return x.reshape(-1, 1), y.reshape(-1, 1)

# Generator model
def build_generator():
    model = models.Sequential([
        layers.Dense(128, activation='relu', input_shape=(100,)),
        layers.Dense(256, activation='relu'),
        layers.Dense(512, activation='relu'),
        layers.Dense(256, activation='relu'),
        layers.Dense(128, activation='relu'),
        layers.Dense(1)
    ])
    return model

# Discriminator model
def build_discriminator():
    model = models.Sequential([
        layers.Dense(128, activation='relu', input_shape=(1,)),
        layers.Dense(256, activation='relu'),
        layers.Dense(512, activation='relu'),
        layers.Dense(256, activation='relu'),
        layers.Dense(128, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])
    return model

# GAN model
def build_gan(generator, discriminator):
    discriminator.trainable = False
    model = models.Sequential([
        generator,
        discriminator
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy')
    return model
```

```python
def train_gan(generator, discriminator, gan, x_train, epochs=1000, batch_size=128, dis_learning_rate=0.001, gen_learning_rate=0.001):
    # Define optimizers for discriminator and generator
    dis_optimizer = tf.keras.optimizers.Adam(learning_rate=dis_learning_rate)
    gen_optimizer = tf.keras.optimizers.Adam(learning_rate=gen_learning_rate)
    
    d_loss_all = []
    g_loss_all = []
    
    for epoch in range(epochs):
        d_loss_total = 0
        g_loss_total = 0
        num_batches = len(x_train) // batch_size
        if num_batches == 0:
            print("Error: Not enough data for the batch size.")
            return [], []  # Return default empty lists
        for _ in range(num_batches):
            # Train discriminator
            noise = np.random.randn(batch_size, 100)
            generated_data = generator.predict(noise)
            real_data = x_train[np.random.randint(0, len(x_train), size=batch_size)]
            X = np.concatenate([real_data, generated_data])
            y_dis = np.zeros(2*batch_size)
            y_dis[:batch_size] = 0.9
            discriminator.trainable = True
            with tf.GradientTape() as tape:
                d_loss = discriminator.train_on_batch(X, y_dis)
            d_loss_total += d_loss

            # Train generator
            noise = np.random.randn(batch_size, 100)
            y_gen = np.ones(batch_size)
            discriminator.trainable = False
            with tf.GradientTape() as tape:
                g_loss = gan.train_on_batch(noise, y_gen)
            g_loss_total += g_loss
        
        # Calculate average losses
        d_loss_avg = d_loss_total / num_batches
        g_loss_avg = g_loss_total / num_batches
        
        # Append to lists for plotting
        d_loss_all.append(d_loss_avg)
        g_loss_all.append(g_loss_avg)
        
        # Print epoch statistics
        if epoch % 100 == 0:
            print(f"Epoch: {epoch}, D Loss: {d_loss_avg}, G Loss: {g_loss_avg}")
    
    # Return the lists of losses for visualization
    return d_loss_all, g_loss_all
```

```python
# Generate synthetic sinusoidal data
num_samples = 1000
x_train, y_train = generate_data(num_samples)

# Normalize data to [-1, 1] range
y_train = (y_train - np.min(y_train)) / (np.max(y_train) - np.min(y_train)) * 2 - 1

# Build generator and discriminator
generator = build_generator()
discriminator = build_discriminator()

# Build and compile GAN
gan = build_gan(generator, discriminator)

# Compile the discriminator and generator separately
discriminator.compile(optimizer='adam', loss='binary_crossentropy')
generator.compile(optimizer='adam', loss='binary_crossentropy')

# Train GAN
d_loss_all, g_loss_all = train_gan(generator, discriminator, gan, x_train, epochs=1000, batch_size=128, dis_learning_rate=0.01, gen_learning_rate=0.01)


# Plot the loss curves
plt.plot(d_loss_all, label='Discriminator Loss')
plt.plot(g_loss_all, label='Generator Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('GAN Training Loss')
plt.legend()
plt.show()
```
