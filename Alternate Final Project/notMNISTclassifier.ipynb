```python
from sklearn.model_selection import train_test_split

x_all = np.array(images_array)
y_all = np.array(labels)

print("x_all shape:", x_all.shape)
print("y_all shape:", y_all.shape)

# Normalize pixel values to the range [0, 1]
x_all = x_all / 255.0

# Split the dataset into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x_all, y_all, test_size=0.2, random_state=42)

# Print the shapes of the resulting datasets
print("x_train shape:", x_train.shape)
print("y_train shape:", y_train.shape)
print("x_test shape:", x_test.shape)
print("y_test shape:", y_test.shape)
```

```python
# Prepare training data
shape = x_train.shape

# Assign dimensions individually
samples = shape[0]
height = shape[1]
width = shape[2]

# Prepare test data
shape = x_train.shape

# Prepare testing data
shape = x_test.shape

# Assign dimensions individually
samples = shape[0]
height = shape[1]
width = shape[2]

X_test = np.reshape(x_test,(samples,width*height))
```

```python
# Import
from sklearn.linear_model import LogisticRegression

# Instantiate
lg = LogisticRegression(multi_class='multinomial', solver='lbfgs', random_state=42, verbose=1, max_iter=1000, n_jobs=-1)

# Fit
lg.fit(X_train, y_train)

# Predict
y_pred = lg.predict(X_test)

# Score
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

report = classification_report(y_test, y_pred)

print(report)

# Plot confusion matrix
xlabels = ['A','B','C','D','E','F','G','H','I','J']
ylabels = xlabels

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='g', xticklabels=xlabels, yticklabels=ylabels)
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```
