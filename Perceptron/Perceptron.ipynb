```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Generate Synthetic Data
np.random.seed(10)
num_samples = 100
# Generate two clusters of data points with different means
class1 = np.random.normal(loc=[3, 3], scale=1, size=(num_samples, 2))
class2 = np.random.normal(loc=[6, 6], scale=1, size=(num_samples, 2))

# Concatenate the data points and create labels
X = np.concatenate([class1, class2])
y = np.concatenate([np.zeros(num_samples), np.ones(num_samples)])

# Step 2: Train Perceptron
from sklearn.linear_model import Perceptron

# Create and train the perceptron model
perceptron = Perceptron()
perceptron.fit(X, y)

# Step 3: Train Logistic Regression
from sklearn.linear_model import LogisticRegression

# Create and train the logistic regression model
logistic_regression = LogisticRegression()
logistic_regression.fit(X, y)

# Step 4: Visualize Decision Boundaries
# Create a mesh grid of points to plot the decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))
```

```python
# Plot the decision boundaries
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot Perceptron decision boundary
Z_perceptron = perceptron.predict(np.c_[xx.ravel(), yy.ravel()])
Z_perceptron = Z_perceptron.reshape(xx.shape)
ax1.contourf(xx, yy, Z_perceptron, cmap=plt.cm.Paired, alpha=0.8)
ax1.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
ax1.set_title('Perceptron Decision Boundary')

# Plot Logistic Regression decision boundary
Z_logistic = logistic_regression.predict(np.c_[xx.ravel(), yy.ravel()])
Z_logistic = Z_logistic.reshape(xx.shape)
ax2.contourf(xx, yy, Z_logistic, cmap=plt.cm.Paired, alpha=0.8)
ax2.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
ax2.set_title('Logistic Regression Decision Boundary')

plt.show()
```
